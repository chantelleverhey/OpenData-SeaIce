# Time Series Sea-Ice Extent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Read the CSV file into a DataFrame - since the file is in the same directory, all you need is the file name
df = pd.read_csv('2008_2023c_sio.csv')
```

```python
# View the first 5 rows of your file
df.head()
```

```python
# Step 4: Plot the time series graph - REMINDER: Jupyter notebook is case sensitive
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['extent'], marker='o', linestyle='-')
plt.title('Sea Ice Outlook 2008-2023 (Pan-arctic region)')
plt.xlabel('Year')
plt.ylabel('Sea Ice Extent')
plt.grid(True)

# Step 4: Fit a line of best fit to the data
slope, intercept, r_value, p_value, std_err = stats.linregress(df['year'], df['extent'])
line = slope * df['year'] + intercept

# Step 5: Plot the line of best fit
plt.plot(df['year'], line, color='red', label='Line of Best Fit')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
